```python
import matplotlib.pyplot as plt

study_hours = [0, 1, 2, 3, 4, 5]
test_scores = [50, 65, 75, 85, 90, 95]

# Create plot with markers for each data point
plt.plot(study_hours, test_scores, marker='o')

# Add descriptive labels
plt.xlabel('Study Hours')
plt.ylabel('Test Score (%)')
plt.title('Impact of Study Time on Test Performance')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert times to 24-hour format for plotting
times = [6, 9, 12, 15, 18, 21]  # 6am to 9pm
temperatures = [15, 18, 22, 25, 21, 17]  # Example temperatures in °C

# Create the plot with a connected line and markers
plt.plot(times, temperatures, marker='o', linestyle='-', linewidth=2)

# Customize the plot
plt.xlabel('Time of Day (24-hour)')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperature Variation')

# Set x-axis ticks to show times clearly
plt.xticks(times, ['6am', '9am', '12pm', '3pm', '6pm', '9pm'])

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Save in both formats
plt.savefig('temperature_plot.png', dpi=300, bbox_inches='tight')
plt.savefig('temperature_plot.pdf', bbox_inches='tight')

plt.show()
```
